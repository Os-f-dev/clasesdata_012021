```python
data_phone = pd.read_csv('data/phone_data.csv')
```

```python
data_phone.head()
```

```python
data_phone.drop('index',axis = 1, inplace = True)
```

```python
data_phone.set_index(['network','item'])['duration']
```

```python
data_phone.groupby(['network','item']).sum()
```

```python
data_phone['network'].value_counts()
```

```python
data_phone.groupby(['month']).mean()
```

```python
data_phone.dtypes
```

```python
data_month = data_phone.groupby(['month','network']).mean()
data_month
```

```python
data_month.unstack()
```

```python
data_phone.groupby(['item','network_type'])['duration'].mean()
```

```python
data_phone.groupby(['item','network_type']).count()
```

```python
data_phone[data_phone['item']== 'call']
```

```python
data_phone[data_phone['item']== 'call'].groupby('month').agg(
            max_d = pd.NamedAgg(column = 'duration', aggfunc = max),
            min_d = pd.NamedAgg(column = 'duration', aggfunc = min),
            total = pd.NamedAgg(column = 'duration', aggfunc = sum))
```

```python
aggregations = {
    # work on the "duration" column
    'duration': { 
        # get the sum, and call this result 'total_duration'
        'total_duration': 'sum',  
        # get mean, call result 'average_duration'
        'average_duration': 'mean', 
        'num_calls': 'count'
    },
    
     # Calculate two results for the 'network' column with a list
    'network': ["count", "max"] 
}
# Perform groupby aggregation by "month", 
# but only on the rows that are of type "call"
data_phone[data_phone['item'] == 'call'].groupby('month').agg(aggregations)
```

```python
carros = pd.read_csv('data/vehicles.csv')
```

```python
carros.head()
```

```python
carros['City MPG'].corr(carros['Highway MPG'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = "white")
corr = carros.corr(method = 'spearman')
mask = np.triu(np.ones_like(corr, dtype = np.bool))
```

```python
mask
```

```python
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(150, 100, as_cmap=True)
sns.heatmap(corr, cmap=cmap, annot=True, mask = mask)
```
