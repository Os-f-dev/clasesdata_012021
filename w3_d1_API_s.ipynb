```python
def flatten(data, cols):
    for c in cols:
        fl = pd.DataFrame(dict(data[c])).T
        cols = [str(col) for col in fl.columns]
        fl.columns = [c+''+name for name in cols]
        data = pd.concat([data,fl], axis = 1)
        data = data.drop(c, axis = 1)
    return data
```

```python
for url in ['https://api.github.com', 'https://api.github.com/invalid']:
    try:
        response = requests.get(url)
        #si la conexion es incorrecta lanzara un error
        response.raise_for_status() #checa el estatus de la respuesta, buscamos 200 si no ira a buscar el error
    
    except HTTPError as http_err:
        print(f'Hubo un error, el cual es : {http_err}')
        #print(url)
    except Exception as err:
        print(f'Ocurrio un segundo error el cual es {err}')
    else:
        print('Todo salio bien')
        print(url)
```
